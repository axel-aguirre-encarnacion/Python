# Análisis Exploratorio de Datos

## Analisis de Variables Númericas


### Crear indicadores derivados

* Total_Treatment: Intensidad total de tratamiento, algunos pacientes reciben ambos tipos de terapia, y tener un indicador conjunto permite comparar la carga total de tratamiento entre individuos.

* Ratio_Tumor_Age: agresividad relativa, Valores altos asociados a tumor grande en un paciente joven y valores bajos asociados a tumor pequeño en un paciente mayor



```python
data['Total_Treatment'] = data['ChemotherapySessions'] + data['RadiationSessions']
data['Ratio_Tumor_Age'] = data['TumorSize'] / data['Age']
```

###Histograma

```python
data.hist(figsize=(12,8), bins=30, color='skyblue', edgecolor='black')
plt.suptitle("Distribución de variables numéricas (pacientes)", fontsize=16)
plt.tight_layout()
plt.show()
```

**Conclusiones por variable**

1. **Edad (`Age`)**  
   Pacientes entre 18 y 85 años, con representación en todo el rango adulto, lo que puede influir en la supervivencia  
   Distribución uniforme, no normal.  

2. **Tamaño del tumor (`TumorSize`)**  
   Promedio de 6–7 cm, la mayoría entre 4 y 10 cm.  
   Distribución cercana a normal, con posibles diagnósticos en etapas intermedias.  

3. **Meses de seguimiento (`FollowUpMonths`)**  
   Varía de 1 a 60 meses, con un promedio de ~30.  
   Distribución uniforme, con pacientes de corto y largo seguimiento.  

4. **Tratamientos (quimio, radio y total)**  
   La mayoría recibió pocos o ningún tratamiento, mientras que solo unos pocos acumularon muchos.  
   Distribución muy sesgada a la derecha, no normal.  

5. **Relación tumor/edad (`Ratio_Tumor_Age`)**  
   La mayoría tiene valores bajos, pero algunos jóvenes con tumores grandes resaltan como casos de riesgo.  
   Distribución sesgada a la derecha.  

###Correlaciones entre variables numéricas

###Matriz de correlacion

En este análisis se aplicaron dos métodos de correlación distintos: **Pearson** y **Kendall**.  
La elección de ambos se debe a las características de los datos clínicos y al interés en obtener una visión más completa de las relaciones entre variables.

```python
df_num = data.select_dtypes(include='number')
cor_matrix = df_num.corr()
cor_matrix_kendall = df_num.corr(method="kendall")

cor_matrix
```

###Mapa de calor

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(10,8))
sns.heatmap(cor_matrix, annot=True, vmin=-1, vmax=1, cmap='RdBu', linewidths=0.5)
plt.title("Mapa de calor - Pearson")
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(cor_matrix_kendall, annot=True, vmin=-1, vmax=1, cmap='RdBu', linewidths=0.5)
plt.title("Mapa de calor - Kendall")
plt.show()
```

**Conclusiones de correlaciones**

El análisis de correlación con Pearson y Kendall muestra resultados consistentes en las relaciones más importantes:

- Existe una fuerte correlación negativa entre **Edad** y **Ratio_Tumor_Age**, lo que refleja que los pacientes más jóvenes con tumores grandes tienen un mayor riesgo relativo.  
- El **Tamaño del tumor** se asocia de forma moderada y positiva con el **Ratio_Tumor_Age**, lo que confirma su influencia directa en ese indicador.  
- Tanto quimio como radio tienen correlaciones positivas claras con el **Total_Treatment**, lo que valida su aporte en la carga total de tratamiento.  
- El **tiempo de seguimiento (FollowUpMonths)** no presenta correlaciones lineales significativas con las demás variables.

**Correlaciones clave:**  
- Las sesiones de quimioterapia y las sesiones de radiación presentan una correlación negativa (-0,27), lo que indica que los pacientes suelen recibir una u otra, no ambas.  
- Otras correlaciones son muy débiles (cercanas a 0), lo que sugiere que las características numéricas son mayoritariamente independientes.

## Analisis de Variables Categoricas


```python
var_categoricos = ['Gender', 'TumorType', 'CancerStage', 'Metastasis', 'TreatmentType', 'SmokingStatus','AlcoholUse','GeneticMutation']
plt.figure(figsize=(18, 12))
for i, col in enumerate(var_categoricos, 1):
    plt.subplot(3, 4, i)
    sns.countplot(data=data, x=col, hue='SurvivalStatus')
    plt.title(f'{col} vs Survival Status')
    plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```

**Conclusiones de Variables Categóricas y Supervivencia**
1. **Género**  
   No se observan diferencias importantes en las proporciones de vivos y fallecidos entre hombres y mujeres, lo que sugiere que el género no es un factor determinante de supervivencia.  

2. **Tipo de Tumor**  
   Los casos de cáncer de pulmón y estómago presentan peores tasas de supervivencia, mientras que los tumores de mama y cervical muestran mejores resultados. Esto indica que el tipo de tumor puede tener un peso moderado en la predicción.  

3. **Estadio del Cáncer**  
   Los pacientes en estadios I y II tienen una mayor supervivencia, mientras que en estadios III y IV la proporción de fallecidos aumenta significativamente. El estadio es uno de los predictores más críticos.  

4. **Metástasis**  
   La ausencia de metástasis se relaciona con una mayor probabilidad de supervivencia, mientras que su presencia incrementa de manera clara la proporción de fallecidos.  

5. **Tipo de Tratamiento**  
   Radiación, quimioterapia y cirugía son los tratamientos más comunes. Aunque las tasas de supervivencia no varían drásticamente entre tipos, se observa una ligera mayor proporción de fallecidos en quienes recibieron radiación o quimioterapia, posiblemente reflejando casos más avanzados.  

6. **Consumo de Tabaco**  
   Los pacientes que nunca fumaron presentan más casos vivos, mientras que los fumadores actuales y exfumadores tienen una mayor proporción de fallecidos. Sin embargo, la diferencia es moderada.  

7. **Consumo de Alcohol**  
   La mayoría de pacientes no presenta consumo de alcohol y se asocia con mayor supervivencia. El consumo excesivo se relaciona con una menor proporción de sobrevivientes.  

8. **Mutaciones Genéticas y Comorbilidades**  
   Los pacientes sin mutaciones ni comorbilidades adicionales presentan mejores resultados de supervivencia. La presencia de mutaciones o enfermedades concomitantes tiende a asociarse con mayores fallecimientos.  

**Conclusión global**  
La supervivencia de los pacientes parece estar principalmente influenciada por el estadio del cáncer y la presencia de metástasis. Otras variables como el tipo de tumor, hábitos de vida o condiciones adicionales muestran impacto, pero en menor medida.

#Recomendación de Metodología de Machine Learning

- **Balance de clases**: aplicar técnicas como SMOTE o class weights, ya que los datos tienen 78% Alive vs 22% Deceased.  

- **Datos faltantes**: manejar variables con vacíos como *GeneticMutation* y *AlcoholUse*, mediante imputación o eliminación si no aportan valor.  

- **Variables sesgadas**: tratar variables como *ChemotherapySessions* y *RadiationSessions* con log-transformación o agrupación en categorías.  

- **Modelos**: comenzar con un modelo interpretable como regresión logística, y luego probar algoritmos más avanzados como Random Forest o XGBoost.  

- **Evaluación**: no limitarse a accuracy. Usar métricas más informativas como AUC, F1 y Recall para evaluar el desempeño en clases desbalanceadas.
