# ANALISIS

##Correlaciones entre variables numéricas

```python
data_prov = data.groupby('Province')[['ChemotherapySessions', 'RadiationSessions', 'TumorSize', 'Age']].mean().reset_index()
```

```python
# Selección de variables numéricas
df_num = data[['Age','TumorSize','ChemotherapySessions','RadiationSessions','FollowUpMonths']]

# Matriz de correlación
cor_matrix = df_num.corr()
cor_matrix

# Mapa de calor
plt.figure(figsize=(8,8))
sns.heatmap(cor_matrix, annot=True, vmin=-1, vmax=1, cmap='RdBu')
plt.title("Mapa de calor - coeficientes de correlación")
plt.show()
```
